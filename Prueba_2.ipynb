## **Ejercicio 1** (40%)

Construya un DataFrame a partir del siguiente diccionario

`exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'], 'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19], 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}`

con las siguientes etiquetas para los índices

`labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']`

1. Seleccione solo las columnas *name* y *score*.
  
2. Filtre las filas donde el número de intentos (attemps) sea mayor que 2.

3. Filtre las filas donde el *score* se encuentre entre 15 y 20 (incluidos)

4. Cambie el *score* en la fila 'd' a 11.5

5. Agregue una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

6. Borre la fila 'k' y entregue nuevamente el DataFrame original.

7. Ordene el DataFrame por nombre en orden ascendente.

8. Reemplace los valores de la columna *qualify* que contiene los valores *yes* y *no* por los valores booleanos True y False respectivamente.

```python
import pandas as pd
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
              'score': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
              'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
              'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

#se construye el data frame con etiquetas desde la letra 'a' a la letra 'j' para el índice
df = pd.DataFrame(exam_data, index=labels)
print(df)
```

1. Seleccionar solo las columnas name y score.

```python
#se seleccionan y se muestran unicamente las columnas name (nombre) y score (puntaje)
s_columnas=df[['name', 'score']]
print(s_columnas)
```

2. Filtrar las filas donde el número de intentos (attemps) sea mayor que 2.

```python
#con el atributo 'loc' se realiza el filtro de la variable attemps que corresponde al numero de intentos, donde este sea mayor a 2
df.loc[df['attempts'] > 2]
```

3. Filtre las filas donde el score se encuentre entre 15 y 20 (incluidos)

```python
#con el atributo 'loc' se realiza el filtro de la variable score que corresponde al puntaje, donde este se encuentre entre 15 y 20 (incluidos)
df.loc[(df.score <= 20) & (df.score >= 15)]
```

4. Cambiar el score en la fila 'd' a 11.5

```python
#con el método 'at' se accede al valor específico que queremos modificar, en este caso 10.0 que será cambiado por 11.5
df.at['d', 'score'] = 11.5
print(df)
```

5. Agregar una nueva fila con etiqueta 'k' al dataframe con un valor que usted desee para cada columna

```python
#con el atributo 'loc' agregamos la nueva fila especificando los valores que se registrarán en ella
df.loc['k'] = ['Laura', 19.5, 2, 'yes']
print(df)
```

6. Borrar la fila 'k' y entregar nuevamente el DataFrame original.

```python
#con el método 'drop' eliminaremos la fila 'k' que fue agregada anteriormente
df = df.drop('k')
print(df)
```

7. Ordenar el DataFrame por nombre en orden ascendente.

```python
#con el método 'sort_values()' y el parámetro 'ascending' se ordena el dataframe por la variable 'name' de forma ascendente (A-Z)
df_ascendente = df.sort_values(by='name', ascending=True)
print(df_ascendente)
```

8. Reemplazar los valores de la columna qualify que contiene los valores yes y no por los valores booleanos True y False respectivamente.

```python
#con el método 'replace' se reemplazan los valores de la varibale 'qualify' por valores booleanos: 'yes' por True y 'no' por False
df['qualify'] = df['qualify'].replace({'yes': True, 'no': False})
print(df)
```

# Solución Ejercicio 2

###1. Se Importa el dataset con el nombre "athletes"

```python
import pandas as pd
# Se importa el dataset athletes.csv
athletes = pd.read_csv('https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/athletes.csv')
athletes
```

###2. Se filtra el dataframe tal que solo se quede con los atletas que hayan ganado al menos una medalla (Bronce, Plata y Oro).

```python
# Se obtienen del DataFrame los atletas que obtuvieron al menos una medalla
athletes[athletes['gold'] + athletes['silver'] + athletes['bronze'] >= 1]
```

###3. Se calcula la masa muscular por fila con una función como

$$I_{m.m.} = \frac{Weight}{Height^2}$$

```python
# Se crea una función ind_masa_musc() que calcula el índice de masa muscular

def ind_masa_musc(row):
  ind_mm = row.weight / (row.height ** 2)
  return ind_mm

athletes.apply(ind_masa_musc, axis = 1)
```

###4. Se agrupan los datos (calculando la media) por deporte para cada una del resto de características. ¿Cuál es el deporte en el que los jugadores tienen mayor peso promedio?

```python
# Se agrupan los datos (calculando la media) por deporte para cada una de las características
athletes.groupby('sport').mean().sort_values(by = 'weight',ascending=False)
```

```python
# Se reduce la vista para mostrar unicamente los deportes y el peso promedio por deporte
athletes.groupby('sport').mean()[['weight']].sort_values(by = 'weight',ascending=False)
```

El **deporte** en el que los jugadores tienen un **mayor peso promedio** es en el **basketball** con un peso medio de **87.75(um)**.

Seguido por balonmano con un preso medio aproximado de 83.71(um)

###5. ¿Cuál es el pais con más participación femenina?

```python
athletes[athletes['sex'] == 'female'].groupby(['nationality','sex']).size().sort_values(ascending = False)
```

El país con mayor participación femenina es **Estados Unidos (USA)** con total de **303 mujeres**.

Seguido por China (CHN) con una participación de 251 mujeres

Importe el dataset de nba del repositorio de clase y responda (hint: Elimine las instancias con registros faltantes):

1. Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)
2. Separe la variable *Age* en tres grupos y cree una nueva columna llamada *Age-Group* con esta información.<
3. Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
import pandas as pd
```

```python
nba_df = pd.read_csv('https://raw.githubusercontent.com/tomasate/Datos_Clases/main/Datos_1/nba.csv')
nba_df
```

```python
nba_df2 = nba_df.dropna()
nba_df2.isnull().values.any()
```

1.Cuáles son los equipos que pagan los salarios más altos (presente el resultado como un dataFrame con el nombre del equipo y la cantidad de dinero de la nómina)

```python
nba_salaries = nba_df2.groupby(['Team']).agg({'Salary':'sum'}).sort_values(by=['Salary'], ascending=False).head()
nba_salaries
```

2.Separe la variable Age en tres grupos y cree una nueva columna llamada Age-Group con esta información

```python
nba_df3 = nba_df2.copy()
nba_df3['Age-Group'] = pd.cut(nba_df2.Age, bins = 3, labels = ['group1', 'group2', 'group3'])
nba_df3
```

3.Construya una tabla dinámica que muestre el salario promedio de los jugadores por equipo y posición para para los cinco equipos que pagan mejores salarios.

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
nba_df3[nba_df3['Team'].isin(nba_salaries.index)].groupby(['Team', 'Position']).agg({'Salary':'mean'})
```
